````python
def getMostPopularSpeechParts(providedDoc):
  notAcceptedSpeechParts = ['NUM','PUNCT','INTJ','SPACE','AUX','SCONJ', 'DET', 'CCONJ','PRON','ADP']
  speechParts = OrderedDict()
  mostPopularSpeechParts = OrderedDict()
  tupledMostPopularSpeechParts = OrderedDict()
  wordsSum = 0
  for token in providedDoc:
    wordsSum += 1
    if token.pos_ not in notAcceptedSpeechParts:
      speechPartList = []
      if token.pos_ in speechParts.keys():
        if speechParts[token.pos_] is None:
          speechParts[token.pos_] = []
        speechPartList.extend(speechParts[token.pos_])
      speechPartList.append(token.text)
      speechParts[token.pos_] = speechPartList
  for index, item in enumerate(speechParts.values()):
    speechPartName = list(speechParts.keys())[index]
    if len(item) > 1:
      counted_dict = dict((i, item.count(i)) for i in item)
      counted_list = [(k, v) for k, v in counted_dict.items()]
      counted_list.sort(key=lambda y: y[1],reverse=True)
      mostPopularSpeechParts[speechPartName] = [k for (k,v) in counted_list]
      tupledMostPopularSpeechParts[speechPartName] = counted_list
  return (mostPopularSpeechParts,tupledMostPopularSpeechParts)
````

````python
def remove_common(a, b):
    # a = b = ["word","word"]
    a, b = list(set(a) - set(b)),list(set(b) - set(a))
    return (a,b)
````

````python
def get_word_from_tuple_list(word,list):
    for tuple in list:
        if tuple[0] == word:
            return tuple
````

````python
def get_list_of_this_word_by_legal_affair(word, dict):
    result = {}
    for legalAffair in dict:
        foundWord = get_word_from_tuple_list(word,dict[legalAffair]['words'])
        if foundWord is not None:
            result[legalAffair] = foundWord
    return result
````

````python
def should_word_be_in_legal_affair(legalAffair, dict):
    numberOfWordInstances = 0
    acceptabilityThreshold = 90
    for legalAffair2 in dict:
        numberOfWordInstances += dict[legalAffair2][1]
    return dict[legalAffair][1]/numberOfWordInstances*100 > acceptabilityThreshold
````

````python
def saveMostPopularWordsWithoutSpeechPartDistinction(dictionary):
    notAcceptedLegalAffairs = ['odpowiedzi niestandardowe','tu interpolska']
    result = OrderedDict()
    for legalAffair in dictionary:
        mergedDictionary = []
        for speechPart in dictionary[legalAffair]:
            mergedDictionary.extend(dictionary[legalAffair][speechPart])
        mergedDictionary.sort(key= lambda x: x[1],reverse=True)
        result[legalAffair] = {'words': mergedDictionary }
    dataFrame = pd.DataFrame(result)
    dataFrame.to_csv("data/result.csv")
    resultWithStrongWords = OrderedDict()
    print('=================')
    print('Accepted speech parts words')
    print('=================')
    for legalAffair in result:
        print(legalAffair+': '+str(len(result[legalAffair]['words'])))
        strongWordsOnly = []
        for word in result[legalAffair]['words']:
            if word[1] >= 2:
                strongWordsOnly.append(word)
        strongWordsOnly.sort(key= lambda x: x[1],reverse=True)
        resultWithStrongWords[legalAffair] = {'words': strongWordsOnly}
    uniqueResult = OrderedDict()
    uniqueResultTuple = OrderedDict()
    print('=================')
    print('Strong words')
    print('=================')
    for legalAffair in resultWithStrongWords:
        print(legalAffair+': '+str(len(resultWithStrongWords[legalAffair]['words'])))
        words = []
        for tuple in resultWithStrongWords[legalAffair]['words']:
            thisWordInDictionary = get_list_of_this_word_by_legal_affair(tuple[0],resultWithStrongWords)
            if len(thisWordInDictionary) == 1 or should_word_be_in_legal_affair(legalAffair, thisWordInDictionary):
                words.append(tuple)
        uniqueResult[legalAffair] = {'words': [*set([k for (k,v) in words])]}
        uniqueResultTuple[legalAffair] = {'words': words}
    legalAffairsList = []
    print('=================')
    print('Unique words')
    print('=================')
    for legalAffair in uniqueResult:
        print(legalAffair+': '+str(len(uniqueResult[legalAffair]['words'])))
        if legalAffair not in notAcceptedLegalAffairs:
            legalAffairsList.append(uniqueResult[legalAffair]['words'])
    with open('data/result.txt', 'w',encoding='utf-8') as f:
        for legalAffair in uniqueResult:
            f.write("\"{}\":{}".format(legalAffair,str(uniqueResult[legalAffair]['words'])))
            f.write('\n')
    f.close()
    with open('data/result-list.txt', 'w',encoding='utf-8') as f:
        f.write(repr(legalAffairsList))
    f.close()
    dataFrame = pd.DataFrame(uniqueResult)
    dataFrame.to_csv("data/unique_result.csv")
    dataFrame = pd.DataFrame(uniqueResultTuple)
    dataFrame.to_csv("data/unique_result_tuples.csv")
````

````python
categories = OrderedDict()

result = OrderedDict()
tupledResult = OrderedDict()
international_docs_sum = 0
constitutional_docs_sum = 0
interpolska_docs_sum = 0
unrecognized_docs_sum = 0
medical_docs_sum = 0
for i in range(len(df)):
  if df['label_high'][i] == 'odpowiedzi niestandardowe':
    unrecognized_docs_sum += 1
    continue
  if df['label_high'][i] == 'tu interpolska':
    interpolska_docs_sum += 1
    continue
  if not df['label_high'][i] in categories:
    categories[df['label_high'][i]]=''
  if df['label_high'][i] == 'prawo konstytucyjne':
    constitutional_docs_sum += 1
  if df['label_high'][i] == 'prawo miädzynarodowe':
    international_docs_sum += 1 
  if df['label_high'][i] == 'prawo medyczne':
    medical_docs_sum += 1 
  categories[df['label_high'][i]] =str(categories[df['label_high'][i]])+" "+str(df['text_full'][i])
print('sum of constitutional docs: {} | {:.2f}%'.format(constitutional_docs_sum,constitutional_docs_sum/2721*100))
print('sum of international docs: {} | {:.2f}%'.format(international_docs_sum,international_docs_sum/2721*100))
print('sum of unrecognized docs: {} | {:.2f}%'.format(unrecognized_docs_sum,unrecognized_docs_sum/2721*100))
print('sum of interpolska docs: {} | {:.2f}%'.format(interpolska_docs_sum,interpolska_docs_sum/2721*100))
print('sum of medical docs: {} | {:.2f}%'.format(medical_docs_sum,medical_docs_sum/2721*100))
print('=================')
print('BEFORE')
print('=================')
for i in categories:
  tmp = categories[i].replace("  "," ").split(" ")
  tmp_set = set(tmp)
  print(i+': '+str(len(tmp_set)) + ' without duplicates and ' + str(len(tmp)) + ' with duplicates')

  # analise the text with pl_core_news_lg package
  analisedText = nlp(categories[i].replace("  "," "))
  print(i + ': ' + str(len(analisedText)))

  # get most popular parts of speech from the document
  doc = getMostPopularSpeechParts(analisedText)[0]
  tupledDoc = getMostPopularSpeechParts(analisedText)[1]
  # save most popular speech parts to the dictionary
  result[i] = doc
  tupledResult[i] = tupledDoc
saveMostPopularWordsWithoutSpeechPartDistinction(tupledResult)

# create a dataframe
dataFrame = pd.DataFrame(result)
# save data to csv
dataFrame.to_csv("data/most_popular_words.csv")
````

````python
def getUniqueDictionary(dictionary):
    speechParts = ['NUM','PUNCT','INTJ','SPACE','AUX','SCONJ', 'DET', 'CCONJ','PRON','ADP']
    uniqueDictionary = OrderedDict()
    for part in speechParts:
        for category in dictionary:
            part1 = dictionary[category]
            for category2 in dictionary:
                part2 = dictionary[category2]
                if category is not category2:
                    if category in uniqueDictionary:
                        part1 = uniqueDictionary[category]
                    if category2 in uniqueDictionary:
                        part2 = uniqueDictionary[category2]
                    if part not in part1.keys():
                        part1[part] = []
                    if part not in part2.keys():
                        part2[part] = []
                    uList1, uList2 = remove_common(part1[part],part2[part])
                    part1[part] = uList1
                    part2[part] = uList2
                    uniqueDictionary[category] = part1
                    uniqueDictionary[category2] = part2
    return uniqueDictionary

def countUniqueWords(dictionary):
    countedDictionary = []
    for part in dictionary:
        values = []
        for words in dictionary[part].values():
            values.append(len(words))
        countedDictionary.append(values)
    return countedDictionary

def reverseDictionary(dictionary):
    speechParts = ['NOUN', 'VERB','ADJ','PROPN']
    reversedDictionary = OrderedDict()
    for part in speechParts:
        partDictionary = OrderedDict()
        for legalAffairs in dictionary:
            parts = dictionary[legalAffairs]
            partDictionary[legalAffairs] = parts[part] if part in parts else []
        reversedDictionary[part] = partDictionary
    return reversedDictionary

````

````python
from sklearn.preprocessing import scale
from minisom import MiniSom
import numpy as np

uniqueDictionary = getUniqueDictionary(result)
reversedDictionary = reverseDictionary(uniqueDictionary)
countedUniqueWords = countUniqueWords(reversedDictionary)

legalAffairs = ['prawo cywilne', 'prawo administracyjne', 'prawo farmaceutyczne','prawo pracy', 'prawo medyczne', 'prawo karne','prawo miädzynarodowe','prawo podatkowe','prawo konstytucyjne']
speechParts = ['NOUN', 'VERB','ADJ','PROPN']

X = scale(countedUniqueWords)
size = 20
som = MiniSom(size, size, len(X[0]),neighborhood_function='gaussian', sigma=1.5,random_seed=1)

som.pca_weights_init(X)
som.train_random(X, len(X)*1000, verbose=True)
````
